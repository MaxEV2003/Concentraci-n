**Para las columnas tarjetas_debito y tarjetas_credito, los valores nulos fueron reemplazados por la mediana debido a la presencia de posibles outliers en los datos.**

Es probable que haya una alta variabilidad en los montos, lo que podría generar valores extremo, los datos de uso de tarjetas pueden variar ampliamente, con algunos clientes realizando compras pequeñas y otros gastando cantidades significativamente mayores.
Los pagos con tarjetas pueden no seguir una distribución simétrica. Por ejemplo, un pequeño porcentaje de transacciones podría ser de montos muy altos, lo que podría sesgar la media.

```python
data["tarjetas_debito"] = data["tarjetas_debito"].fillna(round(data["tarjetas_debito"].median(), 1))
```

```python
valores_nulos = data.isnull().sum()
valores_nulos
```

```python
data["tarjetas_credito"] = data["tarjetas_credito"].fillna(round(data["tarjetas_credito"].median(), 1))
```
